### Dataset
The dataset consists of reviews with their corresponding ratings.  Each data point is a different movie review.  The label, $y$, is $0$ if the review is a bad review and $1$ if the review is a good review.

There are 1000 features, corresponding the the 1000 most common words across all reviews.  For example, the first feature corresponds to the word "the" and the second feature corresponds to the word "and."  The value of the feature is the nubmer of times that word appears in the review.  **Note: This is different from the dataset used for naive Bayes: the features are real-valued. ### Data Pre-processing
The feature values are counts, and therefore different features have different scale.  For instance, the number of times the word "the" appears has a wide range (from 0 to 198 in this dataset), whereas a word like "until" has a much smaller range (from 0 to 5 in this dataset).  For reasons explored in chapters 9 and 12, this can cause problems with learning algorithms.

When used with linear least squares, it was not too much of a problem.  However, with logistic regression, it will cause slow convergence of the method.  With other methods, it can cause poor generalization to the testing data.

Therefore, we will convert each feature from the raw counts to the number of standard deviations this raw count is away from the mean for that feature.  This is called z-score normalization (again, see chapter 9).  We use the
means and standard deviations from the training data, but apply them to both the training and testing data because
this is the sort of information that is "learned" and therefore needs to be based only on the training data.

```python
# z-score normalization:
featuremeans = np.mean(train.X,axis=0)
featurestd = np.std(train.X,axis=0)
train.X = (train.X-featuremeans)/featurestd
test.X = (test.X-featuremeans)/featurestd
```
